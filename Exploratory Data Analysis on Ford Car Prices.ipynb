```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# We can set the style of the plots using seaborn's set_style() function.
# We can choose from different styles:
# - "whitegrid": White background with grid lines
# - "darkgrid": Dark background with grid lines
# - "white": White background without grid lines
# - "dark": Dark background without grid lines
# - "ticks": Traditional plot with ticks on the axes
# The style will be applied to all subsequent plots
sns.set_style("whitegrid")
```

```python
# Creating a count plot of the 'transmission' column
# This plot will show the distribution of different transmission types
# It counts the occurrences of each unique value in the 'transmission' column
sns.countplot(x='transmission', data=df)

# Adding a title to the plot
plt.title('Transmission types')

# Display the plot
plt.show()
```

```python
# Creating a new figure for the pie chart with dimensions 7x7 inches
fig = plt.figure(figsize=(7, 7))

# Define colors for the pie chart slices
colors = ("gold", "red", "green")

# Define wedge properties for the pie chart
wp = {'linewidth': 2, 'edgecolor': "black"}

# Get the counts of each unique value in 'transmission' column
tags = df['transmission'].value_counts()

# Define the amount of "explode" for each slice (separating the slices)
explode = (0.1, 0.1, 0.1)

# Create a pie chart using the 'tags' series directly
tags.plot(kind='pie', autopct='%1.1f%%', shadow=True, colors=colors, startangle=90,
          wedgeprops=wp, explode=explode, label='')

# Adding a title to the pie chart
plt.title('Distribution of the different transmission types')

# Display the pie chart
plt.show()
```

```python
# Create a count plot for fuel types using Seaborn with vertical orientation
sns.countplot(x='fuelType', data=df, palette='viridis')

# Adding a title to the plot
plt.title('Fuel Types')

# Display the plot
plt.show()
```

```python
# Plotting in y axis instead
# Create a count plot for fuel types using Seaborn with vertical orientation
sns.countplot(y='fuelType', data=df, palette='viridis')

# Adding a title to the plot
plt.title('Fuel Types')

# Display the plot
plt.show()
```

```python
# Create subplots with 1 row and 2 columns
fig, axes = plt.subplots(1, 2, figsize=(15, 5), sharey=True)

# Add a title to the entire figure
fig.suptitle('Visualizing Categorical Data Columns')

# Bar plot for 'fuelType' vs. 'price'
sns.barplot(x='fuelType', y='price', data=df, ax=axes[0], palette='viridis')
axes[0].set_title('Fuel Type vs. Price')

# Bar plot for 'transmission' vs. 'price'
sns.barplot(x='transmission', y='price', data=df, ax=axes[1], palette='muted')
axes[1].set_title('Transmission vs. Price')

# Adding labels to the y-axis
axes[0].set_ylabel('Price')
axes[1].set_ylabel('')  # Removing the y-label for the second plot

# Rotating x-axis labels for better readability
axes[0].tick_params(axis='x', rotation=45)
axes[1].tick_params(axis='x', rotation=45)

# Display the plots
plt.tight_layout()
plt.show()
```

```python
###The following code computes the correlation matrix of all numerical columns in the DataFrame `df`. The correlation matrix shows the correlation coefficients between pairs of columns, indicating how strongly they are related. A correlation coefficient close to 1 indicates a strong positive correlation, while a coefficient close to -1 indicates a strong negative correlation.
# Compute the correlation matrix of all numerical columns
df.corr()
```

```python
###The following code creates a heatmap of the correlation matrix of all numerical columns in the DataFrame `df`. The heatmap provides a visual representation of the correlation coefficients between pairs of columns. Annotations are added to the heatmap to display the correlation values in each cell. This helps to identify strong positive or negative correlations between variables.

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 7))  # Set the size of the heatmap
sns.heatmap(df.corr(), annot=True)  # Create the heatmap of the correlation matrix with annotations
plt.title('Correlation between the columns')  # Add a title to the heatmap
plt.show()  # Display the heatmap
```

```python
###The following code calculates the correlation coefficients of all columns with the 'price' column in the DataFrame `df`. It then sorts these correlation values in ascending order, showing how each column is correlated with the car prices. A positive correlation value indicates a direct relationship (as one variable increases, the other also increases), while a negative correlation value indicates an inverse relationship (as one variable increases, the other decreases).
```

```python
# Calculate the correlation of all columns with 'price' and sort
corr_with_price = df.corr()['price'].sort_values()

# Display the sorted correlation values
print(corr_with_price)
```

```python
###The following code creates a regression plot to visualize the correlation between the 'year' and 'price' columns in the DataFrame `df`. The regression plot shows the relationship between the two variables, with 'year' on the y-axis and 'price' on the x-axis. The line in the plot represents the linear regression model fit to the data points, showing the general trend of how car prices change over the years.
# Create a new figure for the regression plot
fig = plt.figure(figsize=(7, 5))

# Set the title for the regression plot
plt.title('Correlation between year and price')

# Create the regression plot
sns.regplot(x='price', y='year', data=df)

# Display the regression plot
plt.show()
```
